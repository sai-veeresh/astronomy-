## FITS file example

```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
from matplotlib.colors import LogNorm
from astropy.visualization import make_lupton_rgb
```

```python
M42u = fits.open('frame-u-006073-4-0063.fits')
M42i = fits.open('frame-i-006073-4-0063.fits')
M42g = fits.open('frame-g-006073-4-0063.fits')
M42r = fits.open('frame-r-006073-4-0063.fits')
M42z = fits.open('frame-z-006073-4-0063.fits') 
```

```python
i = M42i[0].data
u = M42u[0].data
g = M42g[0].data
r = M42r[0].data
z = M42z[0].data
```

```python
print('Data from i band fits:')
print('min:', np.min(i))
print('max:', np.max(i))
print('mean:', np.mean(i))
print('stdev:', np.std(i))
```

```python
print('Data from u band fits')
print('min:', np.min(u))
print('max:', np.max(u))
print('mean:', np.mean(u))
print('stdev:', np.std(u))
```

```python
print('Data from g band fits')
print('min:', np.min(g))
print('max:', np.max(g))
print('mean:', np.mean(g))
print('stdev:', np.std(g))
```

```python
print('Data from r band fits')
print('min:', np.min(r))
print('max:', np.max(r))
print('mean:', np.mean(r))
print('stdev:', np.std(r))
```

```python
print('Data from z band fits')
print('min:', np.min(z))
print('max:', np.max(z))
print('mean:', np.mean(z))
print('stdev:', np.std(z))
```

```python
plt.figure()
plt.imshow(i, cmap='gray')
plt.colorbar()
```

```python
plt.figure(2, figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.imshow(i, cmap='Reds', norm=LogNorm())
plt.subplot(2, 2, 3)
plt.imshow(g, cmap='Greens', norm=LogNorm())
plt.subplot(2, 2, 4)
plt.imshow(u, cmap='Blues', norm=LogNorm())
plt.show()
```

```python
rgb_default = make_lupton_rgb(i, g, u, stretch=1.5, Q=10)
plt.imshow(rgb_default, origin='lower')
plt.show()
```
