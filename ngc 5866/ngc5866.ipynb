```python
import numpy as np 
import matplotlib.pyplot as plt
from astropy.io import fits
from matplotlib.colors import LogNorm
from astropy.visualization import make_lupton_rgb
```

```python
ni=fits.open("frame-i-006122-1-0013.fits")
nu=fits.open("frame-u-006122-1-0013.fits")
ng=fits.open("frame-g-006122-1-0013.fits")
nz=fits.open("frame-z-006122-1-0013.fits")
nr=fits.open("frame-r-006122-1-0013.fits")
```

```python
i=ni[0].data
g=ng[0].data
u=nu[0].data
z=nz[0].data
r=nr[0].data
```

```python
print('for the i band')
ai = type(i)
bi = i.shape
ci = i.dtype.name
print(ai, bi, ci)
print('min:', np.min(i))
print('max:', np.min(i))
print('mean:', np.mean(i))
print('stdev:', np.std(i))
```

```python
print('for the u band')
au = type(u)
bu = u.shape
cu = u.dtype.name
print(au, bu, cu)
print('min:', np.min(u))
print('max:', np.min(u))
print('mean:', np.mean(u))
print('stdev:', np.std(u))
```

```python
print('for the g band')
ag = type(g)
bg = g.shape
cg = g.dtype.name
print(ag, bg, cg)
print('min:', np.min(g))
print('max:', np.min(g))
print('mean:', np.mean(g))
print('stdev:', np.std(g))
```

```python
print('for the r band')
ar = type(r)
br = r.shape
cr = r.dtype.name
print(ar, br, cr)
print('min:', np.min(r))
print('max:', np.min(r))
print('mean:', np.mean(r))
print('stdev:', np.std(r))
```

```python
print('for the z band')
az = type(z)
bz = z.shape
cz = z.dtype.name
print(az, bz, cz)
print('min:', np.min(z))
print('max:', np.min(z))
print('mean:', np.mean(z))
print('stdev:', np.std(z))
```

```python
plt.figure(1, figsize=(10, 5))

plt.subplot(3, 2, 1)
plt.imshow(i, norm=LogNorm())
plt.subplot(3, 2, 2)
plt.imshow(g, norm=LogNorm())
plt.subplot(3, 2, 3)
plt.imshow(u, norm=LogNorm())
plt.subplot(3, 2, 4)
plt.imshow(r, norm=LogNorm())
plt.subplot(3, 2, 5)
plt.imshow(z, norm=LogNorm())
plt.show()
```

```python
rgb_default = make_lupton_rgb(i, r, g, stretch=0.5, Q=10)
plt.imshow(rgb_default, origin='lower')
plt.title('NGC 5866')
plt.show()
```
